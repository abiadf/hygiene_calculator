```python
# all_columns = \
# ['File name', 'Day of measurement', 'Start of post-milk flush [RT]',
#  'Start of pre-rinse [RT]', 'Start of hot rinse [RT]',
#  'Time for max T [RT]', 'Start of post-rinse [RT]',
#  'Start of low-C zone [RT]', 'Duration of low-C zone [s]', 'Max T [C]',
#  'Avg. T of 120s interval with highest T [C]',
#  'Duration for which T>72.0C [s]',
#  'Avg. C for hot rinse (with water) [mS/cm]',
#  'Avg. C for hot rinse (no water) [mS/cm]',
#  'Avg. C for hot rinse (no water) [%]', 'Blowout duration [s]',
#  'Solution type']

relevant_columns = \
['Start of post-milk flush [RT]',
 'Start of pre-rinse [RT]', 'Start of hot rinse [RT]',
 'Time for max T [RT]', 'Start of post-rinse [RT]',
 'Start of low-C zone [RT]', 'Duration of low-C zone [s]', 'Max T [C]',
 'Avg. T of 120s interval with highest T [C]',
 'Duration for which T>72.0C [s]',
 'Avg. C for hot rinse (with water) [mS/cm]',
 'Avg. C for hot rinse (no water) [mS/cm]',
 'Avg. C for hot rinse (no water) [%]', 'Blowout duration [s]',
 'Solution type']

df_relevant = df[relevant_columns].copy()
df_relevant['postmilk'] = pd.to_datetime(df['Start of post-milk flush [RT]'], format='%H:%M:%S')
df_relevant['prerinse'] = pd.to_datetime(df['Start of pre-rinse [RT]'], format='%H:%M:%S')
df_relevant['hotrinse'] = pd.to_datetime(df['Start of hot rinse [RT]'], format='%H:%M:%S')
df_relevant['postrinse']= pd.to_datetime(df['Start of post-rinse [RT]'], format='%H:%M:%S')

df_new = pd.DataFrame()
df_new['postmilk']               = (df_relevant['prerinse'] - df_relevant['postmilk']).dt.total_seconds()
df_new['premilk']                = (df_relevant['hotrinse'] - df_relevant['prerinse']).dt.total_seconds()
df_new['hotrinse']               = (df_relevant['postrinse'] - df_relevant['hotrinse']).dt.total_seconds()
df_new['low C [s]']              = df_relevant['Duration of low-C zone [s]']
df_new['T_max']                  = df_relevant['Max T [C]']
df_new['T of hottest 2min']      = df_relevant['Avg. T of 120s interval with highest T [C]']
df_new['How long T>72']          = df_relevant['Duration for which T>72.0C [s]']
df_new['C_hotrinse (water)']     = df_relevant['Avg. C for hot rinse (with water) [mS/cm]']
df_new['C_hotrinse (no water)']  = df_relevant['Avg. C for hot rinse (no water) [mS/cm]']
df_new['C_hotrinse (no water) %']= df_relevant['Avg. C for hot rinse (no water) [%]']
df_new['blowout [s]']            = df_relevant['Blowout duration [s]']
df_new['Solution type']          = df_relevant['Solution type'].map({'acid': 1, 'alkaline': 0})


# maybe take our y as C_hotrinse (no water)
df_new['Is hygienic'] = 1 # T>72 for 14s OR long hot rinse OR C_hotrinse

df_new
```
